```python
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder

# Re-create X and y after initial label encoding
X = df.drop(columns='Accident Severity')
y = df['Accident Severity']

# Handle missing values in 'Traffic Control Presence' before encoding
if 'Traffic Control Presence' in X.columns:
    mode_tcp = X['Traffic Control Presence'].mode()[0]
    X['Traffic Control Presence'].fillna(mode_tcp, inplace=True)

# Identify categorical columns to encode (excluding 'Accident Severity' which is the target)
categorical_cols = X.select_dtypes(include='object').columns

# Apply Label Encoding to all categorical columns in X
le = LabelEncoder()
for col in categorical_cols:
    X[col] = le.fit_transform(X[col])

X_res, y_res = SMOTE().fit_resample(X, y)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix, classification_report, recall_score, f1_score, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
```

```python
# Split the resampled data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)

# Create a dictionary of models
models = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(n_estimators=10),
    'KNeighbors Classifier': KNeighborsClassifier()
}

results = []
for name, model in models.items():
    model.fit(X_train, y_train)
    y_preds = model.predict(X_test)
    score = accuracy_score(y_test, y_preds)
    pre = precision_score(y_test, y_preds, average='macro')

    print(f"\n{name} Performance:")
    print("Accuracy:", score)
    print("Precision Score:", pre)

    results.append({
        'Model': name,
        'Accuracy': score,
        'Precision': pre
    })

results_df = pd.DataFrame(results)

print("\nModel Evaluation Summary:")
display(results_df)
```

```python
best_model = models['Random Forest']

# Define a placeholder for the new data point
# Replace this with your actual new data in the same format as X_train
new_data_point = X_test.iloc[0].values.reshape(1, -1) # Using the first row of X_test as an example

prediction = best_model.predict(new_data_point)
print(f"The predicted accident severity is: {prediction[0]}")
```
